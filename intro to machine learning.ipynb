```python
#import python library called pandas and abbreviate as pd
import pandas as pd
#import data using pd.read_csv in bracket use r to read the file and inside the "" write the location of the data
titanics_data = pd.read_csv(r"C:\Users\David gathara marigi\Downloads\train.csv")
#data.head print the first five output of the data
titanics_data.head()
```

```python
#the command replaces all missing (NaN) values in the Titanic dataset with 0 and modifies the original DataFrame without creating a copy.
titanics_data.fillna(0, inplace=True)
```

```python
# ways to select a subset of your data
# 1 Dot notation, which we use to select the "prediction target"
y = titanics_data.Survived
# 2 Selecting with a column list, which we use to select the "features"
features = ['PassengerId', 'Age', 'Sex']
X = pd.get_dummies(titanics_data[features])
```

```python
X.describe()
```

```python
X.head()
```

```python
#Here is an example of defining a logistic model with scikit-learn and fitting it with the features and target variable.
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
missing_val_count_by_column = (X_train.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
# Make predictions on the test data
y_pred = model.predict(X_valid)
```

```python
# Evaluate the model's performance
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Accuracy
accuracy = accuracy_score(y_valid, y_pred)
print("Accuracy:", accuracy)

# Classification Report
print(classification_report(y_valid, y_pred))

# Confusion Matrix
print(confusion_matrix(y_valid, y_pred))
```
