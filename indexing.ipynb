```python
'''selecting data based on its numerical position in the data. iloc follows this paradigm. 
To select the first row of data in a DataFrame, we may use the following:'''
home_data.iloc[0]
```

```python
'''Both loc and iloc are row-first, column-second. This is the opposite of what we do in native Python, which is column-first, row-second.
This means that it's marginally easier to retrieve rows, and marginally harder to get retrieve columns. 
To get a column with iloc, we can do the following:'''
print(home_data.iloc[:, 0])
# in the output below we accessed Id column
```

```python
print(home_data.iloc[:, 3])
```

```python
#to select the street column from just the first, second, and third row, we would do:
print(home_data.iloc[0:3, 5])
```

```python
'''it's worth knowing that negative numbers can be used in selection. This will start counting forwards from the end of the values. 
 So for example here are the last five elements of the dataset.'''
print(home_data.iloc[-5:])
```

```python
#Here, :-5 means “select all rows up to (but not including) the fifth-to-last row.”
print(home_data.iloc[:-5])
```

```python
#to get the last entry in home_data, we would now do the following:
print(home_data.loc[0, 'SalePrice'])
```

```python
print(home_data.loc[: ,['MoSold', 'YrSold', 'SalePrice']])
```
